### But how many customers are there from each country? 
* The `GROUP BY` statement groups rows that have the same values into summary rows
* see https://www.w3schools.com/sql/sql_groupby.asp for more information

```python
pd.read_sql("""
            SELECT Country, count(Country)
            FROM customers
            GROUP BY Country
            ORDER BY count(Country) DESC
            """
            , DB) 
```

### List all tracks from the Albums of an Artist


```python
artist = input('Whom are you looking for: ')    # remember the input function?
                                                # and f-Strings?
                                                # why is this generally a bad idea?
df = pd.read_sql(f"""
            SELECT 
                artists.Name as Artist, 
                albums.Title as Album,
                tracks.Name as Track,
                tracks.trackiD
            FROM albums
            INNER JOIN artists
                ON artists.ArtistID = albums.ArtistID
            INNER JOIN tracks
                ON tracks.AlbumID = albums.AlbumID
            WHERE artists.Name LIKE '%{artist}%'
            """, DB)
df
```

### Some side remarks on _Pandas_ `dataframe` (in advance of the sessions after the break)
* the `read_sql` method returns a _Pandas_ `dataframe`
* the data frame is similar to a 2-dimensional table
* _like_ a `list` of `tuples` (the rows) including a generated index at the beginning
* in addition it contains the column heads (labels)

```python
df.head(7) # returns the n first tuples
```

```python
df.tail(7) # returns the n last tuples
```

```python
df['Artist'] # returns the column with label 'string', same as df.Artist
```

```python
df[2:12] # slices the rows
```

```python
# iterating over the returned tuples

for i, artist, album, track, trackId in df.itertuples():
    print(f"Track {i+1} is called '{track}' by {artist} from the album '{album}'")
```

### List the Artists with the most Tracks in the store (Top 10)


```python
pd.read_sql("""
            SELECT 
                artists.Name as Artist, 
                count(tracks.TrackId) as 'Number of tracks'
            FROM albums
            INNER JOIN artists
                ON artists.ArtistID = albums.ArtistID
            INNER JOIN tracks
                ON tracks.AlbumID = albums.AlbumID
            WHERE artists.Name <> 'Various Artists'
            GROUP BY artists.Name
            ORDER BY count(tracks.TrackId) DESC
            LIMIT 10
            """, DB) 

# any improvements? Various Artists?
```

### Who are the best selling Artists in the store (Top 10)?


```python
pd.read_sql("""
            SELECT 
                artists.Name as Artist, 
                count(invoice_items.InvoiceLineId) as 'Number of sales'
            FROM invoice_items
            INNER JOIN tracks
                ON tracks.TrackId = invoice_items.TrackId
            INNER JOIN albums
                ON albums.AlbumID = tracks.AlbumID
            INNER JOIN artists
                ON artists.ArtistID = albums.ArtistID
            GROUP BY artists.Name
            ORDER BY count(invoice_items.InvoiceLineId) DESC
            LIMIT 10
            """, DB) 

# any improvements?
```

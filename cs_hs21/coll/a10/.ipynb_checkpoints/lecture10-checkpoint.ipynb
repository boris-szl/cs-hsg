```python
df_cashflow = df_msft.cashflow
df_cashflow
```

```python
series_operating_cashflow = df_cashflow.loc["Total Cash From Operating Activities"]
series_operating_cashflow = series_operating_cashflow.sort_values(ascending=True)
series_investing_cashflow = df_cashflow.loc["Total Cashflows From Investing Activities"]
series_investing_cashflow = series_investing_cashflow.sort_values(ascending=True)
```

```python
xvals=[2018,2019,2020,2021]
ax1 = plt.subplot(1,2,1)
plt.bar(xvals,series_operating_cashflow)
ax2 = plt.subplot(1,2,2)
plt.bar(xvals,series_investing_cashflow)
```

```python
# we create 2x2 grid of axis subplots
fig, ((ax1,ax2), (ax3,ax4)) = plt.subplots(2,2,sharex=True)
axs = [ax1,ax2,ax3,ax4]
# draw n samples and plot
for n in range(0,len(axs)):
    sample_size = 10**(n+1)
    sample = np.random.normal(loc=0.0, scale=1.0,size=sample_size)
    axs[n].hist(sample,bins=100)
    axs[n].set_title("n={}".format(sample_size))
plt.show()
```

```python
# we create 2x2 grid of axis subplots
fig, ((ax1,ax2), (ax3,ax4)) = plt.subplots(2,2,sharex=True)
axs = [ax1,ax2,ax3,ax4]
# draw n samples and plot
for n in range(0,len(axs)):
    sample_size = 10**(n+1)
    sample = np.random.normal(loc=0.0, scale=1.0,size=sample_size)
    axs[n].hist(sample,bins=10)
    axs[n].set_title("n={}".format(sample_size))
plt.show()
```

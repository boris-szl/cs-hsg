Data scientists spend about 80% of their time cleaning and organizing data. Tidy Data is a method for structuring data sets to facilitate analysis.
In 2014, [Hadley Wickham](https://en.wikipedia.org/wiki/Hadley_Wickham) published a great paper called [Tidy Data](https://vita.had.co.nz/papers/tidy-data.pdf), which describes the process of cleansing a data set.

In tidy data:

1. Each **variable** is saved in its own **column**

variable column

2. Each **obseration** is saved in its own **row**
![observation row]( )

3. Each type of observational unit forms a table.

## Melt

The `melt` function is used to convert wide DataFrames into narrow ones. A wide dataframe has a large number of columns. Some data frames are structured so that successive measurements or variables are displayed as columns. In some cases, displaying these columns as rows may be more appropriate for our task.


![melt]( )

Consider the following dataframe:

```python
import pandas as pd
import numpy as np
df = pd.DataFrame({'city':['A','B','C'],
                   'day1':[22,25,28],
                   'day2':[10,14,13],
                   'day3':[25,22,26],
                   'day4':[18,15,17],
                   'day5':[12,14,18]})
df
```

For three different cities we have carried out measurements on different days. The days were noted as columns of the table. This table does not represent **tidy data** because the column headings, i.e. the days, are values and not variable names. This is against the rule "*Each variable must have its own column*".

To clean up the data, we will display these days as rows in a column.  There will also be a column to represent the measurements, to follow the rule "Each observation must have its own row". Therefore, we aim to have each row in the table as an observation, i.e. a measurement with day and value.

We can easily achieve this by using the `melt` function:

```python
df.melt(id_vars=['city'])
```

variable and value column names are given by default. We can use `var_name` and `value_name` parameters of `melt` function to assign new column names. It will also look better if we sort the data by city column:

```python
df.melt(id_vars=['city'], 
        var_name = 'date', 
        value_name = 'temperature').sort_values(by='city').reset_index(drop=True)
```

## Pivot

The pivot function can be seen as a way to view the dataframe from a different perspective. It is used to examine the relationships between variables by allowing data to be presented in different formats.

![Pivot]( )

Consider the following dataframe:

```python
df = pd.DataFrame(
    {
        'city':['A','B','C','A','C'],
        'name':['John','Tony','Adam','Tony','John'],
        'value1':[10,14,13,9,10],
        'value2':[25,26,18,30,29],
    }
    )
df
```

We want to see how values change according to city-name pairs. We can create a new representation of this dataframe with an index of names and columns of cities.

```python
df.pivot(index='name', columns='city')
```

If a city-name pair does not exist, corresponding cell is filled with 'NaN'.

We do not have to see all the values at once. The values to put in the dataframe can be filtered using values parameter:


```python
df.pivot(index='name', columns='city', values='value1')
```

```python
df.pivot_table(values='(kW)', index='YEAR', columns='Make', aggfunc=[np.mean,len])
```

```python
df.pivot_table(values='(kW)', index='YEAR', columns='Make', aggfunc=[np.mean,np.min], margins=True)
```

```python
class Animal:
    def __init__(self, name, species):
        self.name = name
        self.species = species
        if not isinstance(name, str):
            raise TypeError("Name must be a string!")
            
    def move(self):
        if self.species == "Bird":
            action = "The bird flies away"
        elif self.species == "Fish":
            action = "The fish swims away"
        else:
            action = "The forehead lobe basilisk runs away"
        print(action)
```

```python
bird = Animal("Woodpecker", "Bird")
fish = Animal("Trout", "Fish")
basilisk = Animal("Forehead lobe", "")

fish.move()
```

```python
student_df = pd.DataFrame([{'Name' : 'Thomas', 'School' : 'Computer Science'},
                          {'Name' : 'Reto', 'School' : 'Law'},
                          {'Name' : 'Christian', 'School' : 'Business'}])
students_df = student_df.set_index('Name')

staff_df = pd.DataFrame([{'Name' : 'Myriam', 'Role' : 'Director of HR'},
                          {'Name' : 'Christina', 'Role' : 'Course liasion'},
                          {'Name' : 'Thomas', 'Role' : 'Grader'}])
staff_df = staff_df.set_index('Name')
```

```python
student_df
```

```python
staff_df
```

```python
# vereinigungsmenge zweier datentabellen
pd.merge(student_df, staff_df, how="outer", left_index=True, right_index=True)
```

```python
# schnittmenge zweier datentabellen
pd.merge(student_df, staff_df, how="inner", left_index=True, right_index=True)
```

```python
# use only keys from right frame, similar to a SQL right outer join; preserve key order.
pd.merge(student_df, staff_df, how="right", left_index=True, right_index=True)
```

```python
# uses only keys from left frame, similar to a SQL left outer join; preserve key order.
pd.merge(student_df, staff_df, how="left", left_index=True, right_index=True)
```

```python
pd.merge(student_df, staff_df, how="left",left_on="Name", right_on="Name")
```

```python
student_df.reset_index()
staff_df.reset_index()
pd.merge(student_df, staff_df, how="outer", left_on="Name", right_on="Name")
```
